# Ch.3 데이터 분석 맛보기

```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome()

url = 'http://www.melon.com/chart/index.htm'
driver.get(url)

soup = BeautifulSoup(driver.page_source, 'html.parser')

song_data = []
songs = soup.select('table > tbody > tr')

for rank, song in enumerate(songs):
    title = song.select('div.rank01 > span > a')[0].text
    singer = song.select('div.rank02 > span > a')[0].text
    song_data.append(['Melon', rank+1, title, singer])
```

```python
# 코드 간결하게 변경 (결과는 song_data와 동일)
new_song_data = [
    [
        'Melon', rank+1,
        song.select('div.rank01 > span > a')[0].text,
        song.select('div.rank02 > span > a')[0].text
    ] for rank, song in enumerate(songs)
    ]
new_song_data
```

```python
#  위의 예제 데이터 프레임으로 만들기 (데이터 정리)
import pandas as pd

columns = ['Service', 'Rank', 'Title', 'Singer']
df_song = pd.DataFrame(song_data, columns=columns)
df_song.head()
```

```python
df_song.to_excel('C:/Users/ChangHo Kim/Documents/GitHub/SSAC_study/WebCrawling/Sample/melon_song_100.xlsx', index = False)
```

## 벅스 크롤링

```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome()

url = 'https://music.bugs.co.kr/chart'
driver.get(url)

soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
songs = soup.select('table.byChart > tbody > tr')
song = songs[0]

bugs_song_data = [
    [
        'Bugs', rank+1,
        song.select('p.title')[0].text.strip(),
        song.select('p.artist')[0].text.strip()
    ] for rank, song in enumerate(songs)
    ]
bugs_song_data
```

```python
import pandas as pd

columns = ['Service', 'Rank', 'Title', 'Singer']
df_bugs_song = pd.DataFrame(bugs_song_data, columns=columns)
df_bugs_song.head()
```

```python
df_bugs_song.to_excel('C:/Users/ChangHo Kim/Documents/GitHub/SSAC_study/WebCrawling/Sample/bugs_song_100.xlsx', index = False)
```

## 지니 음악 Top 50

```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome()

url = 'https://www.genie.co.kr/chart/top200'
driver.get(url)

soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
songs = soup.select('table > tbody > tr')
song = songs[0]

g_song_data = [
    [
        '지니', rank+1,
        song.select('a.title')[0].text.strip(),
        song.select('a.artist')[0].text.strip()
    ] for rank, song in enumerate(songs)
    ]

g_song_data
```

```python
import pandas as pd

columns = ['Service', 'Rank', 'Title', 'Singer']
df_g_song = pd.DataFrame(g_song_data, columns=columns)
df_g_song.head()
```

```python
df_g_song.to_excel('C:/Users/ChangHo Kim/Documents/GitHub/SSAC_study/WebCrawling/Sample/g_song_50.xlsx', index = False)
```

## 멜론, 벅스, 지니 크롤링 엑셀 파일 합치기

```python
excel_file_names = ['melon_song_100' , 'bugs_song_100', 'g_song_50']
l = len(excel_file_names)
df_final = pd.DataFrame()

for names in zip(['./Sample/']*l , excel_file_names, ['.xlsx']*l):
    name = ''.join(names)
    df_songs = pd.read_excel(name)
    df_final = df_final.append(df_songs)
df_final
```
