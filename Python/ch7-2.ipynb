```python
from urllib import request
from bs4 import BeautifulSoup

url = "https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
target = request.urlopen(url)

soup = BeautifulSoup(target, "html.parser")

for location in soup.select("location"): # select : 나오는 모든 것을 찾음
    print("City:", location.select_one("city").string) # select_one : 처음나오는 "city" 값 반환
    print("Weather:", location.select_one("wf").string)
    print("Min. Temp.:", location.select_one("tmn").string)
    print("Max. Temp.:", location.select_one("tmx").string, "\n")
```

```python
from flask import Flask
from urllib import request
from bs4 import BeautifulSoup

# Web server
url = "https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
app = Flask(__name__)
@app.route("/")

def hello():
    
    target = request.urlopen(url)
    
    soup = BeautifulSoup(target, "html.parser")
    
    output = ""
    for location in soup.select("location"):
        output += "<h3>{}</h3>".format(location.select_one("city").string)
        output += "날씨: {}<br>".format(location.select_one("wf").string)
        output += "최저/최고 기온: {}/{}"\
            .format(\
                location.select_one("tmn").string,
                location.select_one("tmx").string)
        output += "<hr/>"
    return output

# 바탕화면에 app,py로 구현완료함
# cmd에서 flask run 통해 실행
```

```python
# 함수 데코레이터
def test(function):
    def wrapper():
        print("hi")
        function()
        print("bye")
    return wrapper

@test # 데코레이터를 통해서 함수를 자동호출
def hello():
    print("hello")

hello()
```
