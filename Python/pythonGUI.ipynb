```python
import tkinter as tk
import tkinter.font as font

class App:
    def __init__(self):
        root = tk.Tk()

        self.customFont = font.Font(family='Arial', size=12)

        buttonframe = tk.Frame()
        label = tk.Label(root, text='Hello, World!', font=self.customFont)
        buttonframe.pack()
        label.pack()

        bigger = tk.Button(root, text='size up', command=self.BigFont)
        smaller = tk.Button(root, text='size down', command=self.SmallFont)
        bigger.pack()
        smaller.pack()
        root.mainloop()
    
    def BigFont(self):
        size = self.customFont['size']
        self.customFont.configure(size=size+2)

    def SmallFont(self):
        size = self.customFont['size']
        self.customFont.configure(size=size-2)

app = App()
```

```python
from tkinter import *
w = 500
h = 300

def drawDot(event):
    x1, y1 = (event.x-1), (event.y-1)
    x2, y2 = (event.x+1), (event.y+1)
    canvas.create_oval(x1, y1, x2, y2, fill='white')

window = Tk()
canvas = Canvas(window, width=w, height=h)
canvas.pack(expand=YES, fill=BOTH)
canvas.bind('<B1-Motion>', drawDot)

message = Label(window, text='마우스 드래그 시 점이 그려짐')
message.pack(side=BOTTOM)

window.mainloop()
```

```python
# 타이머 초기화 버튼 추가하기 (timer=0 만드는 함수 만들기)
import tkinter as tk

def startTimer():
    if (running):
        global timer
        timer += 1
        timeText.configure(text=str(timer))
    window.after(10, startTimer)

def start():
    global running
    running = True

def stop():
    global running
    running = False

def reset():
    global timer
    timer = 0

running = False
window = tk.Tk()

timer = 0

timeText = tk.Label(window, text='0', font=('Arial', 80))
timeText.pack()

startButton = tk.Button(window, text='start', bg='yellow', command=start)
startButton.pack(fill=tk.BOTH)

stopButton = tk.Button(window, text='stop', bg='yellow', command=stop)
stopButton.pack(fill=tk.BOTH)

resetButton = tk.Button(window, text='reset', bg='yellow', command=reset)
resetButton.pack(fill=tk.BOTH)

startTimer()
window.mainloop()
```

```python
# 계산기 만들기
from tkinter import *

def click(key):
    if key == '=': # '=' 버튼으로 수식을 계산
        try:
            result = eval(entry.get())
            entry.delete(0, END)
            entry.insert(END, str(result))
        except:
            entry.insert(END, 'ERROR!')
    
    elif key == 'C':
        entry.delete(0, END)

    else:
        entry.insert(END, key)

window = Tk()
window.title('Simple Calculator')
buttons = [
    '7', '8', '9', '+', 'C',
    '4', '5', '6', '-', ' ',
    '1', '2', '3', '*', ' ',
    '0', '.', '=', '/', ' '
]

# 반복문으로 버튼을 생성
i = 0
for b in buttons:
    cmd = lambda x=b: click(x)
    b = Button(window, text=b, width=5, relief='ridge', command=cmd)
    b.grid(row=i//5+1, column=i%5)
    i += 1

# 엔트리 위젯은 맨 윗줄의 5개의 셀에 걸쳐서 배치된다
entry = Entry(window, width=33, bg='yellow')
entry.grid(row=0, column=0, columnspan=5)

window.mainloop()
```
